## Convert date time string to number

```python
new_data["Initiation Time"]=pd.to_datetime(new_data["Initiation Time"],format="%d-%m-%Y %H:%M:%S" , errors='coerce')
```

```python
new_data.info()
```

## New date and time format

```python
new_data['hour'] = new_data['Initiation Time'].dt.hour
new_data['minute'] = new_data["Initiation Time"].dt.minute
new_data['seconds'] = new_data['Initiation Time'].dt.second
new_data['date'] = new_data["Initiation Time"].dt.date
new_data['time'] = new_data["Initiation Time"].dt.time
```

```python
new_data
```

```python
hours = [hour for hour, df in new_data.groupby("hour")]

plt.plot(hours, new_data.groupby(['hour']).count())
plt.xticks(hours)
plt.xlabel("Hour")
plt.ylabel("Number of Transcations")
plt.grid()
plt.show()
```
